## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x1 = np.arange(-5, -0.04, 0.05)
x2 = np.arange(0.04, 5, 0.05)
x = np.array(x1+x2)


def f(x, a, b):
    return (x**b+a**b)/x**b


#часть 1
fig, axes = plt.subplots(figsize = (10, 10))
plt.ylim(-20, 30) 
plt.xlim(-5, 5)
axes.plot(x1, f(x1, 1, 1), 'r')
axes.plot(x2, f(x2, 1, 1), 'r', label=r'$\alpha=1,\beta=1$')
axes.plot(x1, f(x1, 2, 1), 'b')
axes.plot(x2, f(x2, 2, 1), 'b', label=r'$\alpha=2,\beta=1$')
axes.plot(x1, f(x1, 1, 2), 'g')
axes.plot(x2, f(x2, 1, 2), 'g', label=r'$\alpha=1,\beta=2$')
axes.set_xlabel('$𝑥$', fontsize = 30)
axes.set_ylabel('$𝑓(𝑥)$', fontsize = 30)
axes.set_title('График функции '  r'$𝑓(𝑥) =\frac{𝑥^\beta+\alpha^\beta}{𝑥^\beta}$' ' при различных ' r'$\alpha$' ' и ' r'$\beta$', fontsize = 30)
axes.legend(fontsize = 16)
plt.savefig('1-1.svg')
```

### Часть 2

Построить в общих осях графики для $x>0$.

На том же графике сделать 2 врезки, демонстрирующие поведение графиков на 2 интервалах:

- для малых x

- для больших x

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
#Часть 2
def zero(x):
    return x*0
fig1 = plt.figure(figsize = (10, 10))

axes1 = fig1.add_axes([0, 0, 1, 1])
axes1.plot(x2, f(x2, 1, 1), 'r', label=r'$\alpha=1,\beta=1$')
axes1.plot(x2, f(x2, 2, 1), 'b', label=r'$\alpha=2,\beta=1$')
axes1.plot(x2, f(x2, 1, 2), 'g', label=r'$\alpha=1,\beta=2$')
axes1.plot(x2, zero(x2), 'k', label=r'$𝑥 = 0$')
axes2 = fig1.add_axes([0.3, 0.7, 0.4, 0.2])
plt.xlim(0, 0.5)

axes2.plot(x2, f(x2, 1, 1), 'r', label=r'$\alpha=1,\beta=1$')
axes2.plot(x2, f(x2, 2, 1), 'b', label=r'$\alpha=2,\beta=1$')
axes2.plot(x2, f(x2, 1, 2), 'g', label=r'$\alpha=1,\beta=2$')
axes2.plot(x2, zero(x2), 'k', label=r'$𝑥 = 0$')
axes3 = fig1.add_axes([0.3, 0.3, 0.4, 0.2])
plt.xlim(0.5, 5)
axes3.plot(x2, f(x2, 1, 1), 'r', label=r'$\alpha=1,\beta=1$')
axes3.plot(x2, f(x2, 2, 1), 'b', label=r'$\alpha=2,\beta=1$')
axes3.plot(x2, f(x2, 1, 2), 'g', label=r'$\alpha=1,\beta=2$')
axes3.plot(x2, zero(x2), 'k', label=r'$𝑥 = 0$')

    
axes1.set_xlabel('$𝑥$', fontsize = 30)
axes1.set_ylabel('$𝑓(𝑥)$', fontsize = 30)
axes1.set_title('График функции '  r'$𝑓(𝑥) =\frac{𝑥^\beta+\alpha^\beta}{𝑥^\beta}$' 'при ' r'$𝑥$''>0', fontsize = 30)
axes1.legend(fontsize = 16)
axes2.set_xlabel('$𝑥$', fontsize = 16)
axes2.set_ylabel('$𝑓(𝑥)$', fontsize = 16)
axes2.set_title('Поведение функции при малых' r'$𝑥$', fontsize = 16)
axes2.legend(fontsize = 10)
axes3.set_xlabel('$𝑥$', fontsize = 16)
axes3.set_ylabel('$𝑓(𝑥)$', fontsize = 16)
axes3.set_title('Поведение функции при больших' r'$𝑥$', fontsize = 16)
axes3.legend(fontsize = 10)
    
plt.savefig('1-2.svg')
plt.show()
```

### Часть 3

Построить в общих осях графики для $x<0$.

На том же графике сделать 1 врезку, демонстрирующую поведение графиков при удалении x от 0 к $-\infty$.

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций. Так же нанесите на графики прямую f(x) = 0.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
#Часть 3 

fig2 = plt.figure(figsize = (10, 10))

axes4 = fig2.add_axes([0, 0, 1, 1])
axes4.plot(x1, f(x1, 1, 1), 'r', label=r'$\alpha=1,\beta=1$')
axes4.plot(x1, f(x1, 2, 1), 'b', label=r'$\alpha=2,\beta=1$')
axes4.plot(x1, f(x1, 1, 2), 'g', label=r'$\alpha=1,\beta=2$')
axes4.plot(x1, zero(x2), 'k', label=r'$𝑥 = 0$')
axes5 = fig2.add_axes([0.3, 0.7, 0.6, 0.2])
plt.xlim(-5, -0.5)
axes5.plot(x1, f(x1, 1, 1), 'r', label=r'$\alpha=1,\beta=1$')
axes5.plot(x1, f(x1, 2, 1), 'b', label=r'$\alpha=2,\beta=1$')
axes5.plot(x1, f(x1, 1, 2), 'g', label=r'$\alpha=1,\beta=2$')
axes5.plot(x1, zero(x2), 'k', label=r'$𝑥 = 0$')

    #Легенды
    
axes4.set_xlabel('$𝑥$', fontsize = 30)
axes4.set_ylabel('$𝑓(𝑥)$', fontsize = 30)
axes4.set_title('График функции '  r'$𝑓(𝑥) =\frac{𝑥^\beta+\alpha^\beta}{𝑥^\beta}$' 'при ' r'$𝑥$''<0', fontsize = 30)
axes4.legend(fontsize = 16)
axes5.set_xlabel('$𝑥$', fontsize = 16)
axes5.set_ylabel('$𝑓(𝑥)$', fontsize = 16)
axes5.set_title('Поведение функции при' r'$𝑥$''стремщемся к −∞', fontsize = 16)
axes5.legend(fontsize = 10)
    
plt.savefig('1-3.svg')
plt.show()
print(u'Выводы:')
print(u'1. Изменение коэффициента α приводит к смещению графика функции по оси x')
print(u'2. По мере удаления абсциссы от нуля функции стремятся к нулю')
print(u'3. По мере приближения абсциссы к нулю функции стремятся к +- бесконечности, что зависит от чётности коэффициента β')
print(u'4. Функции убывают на интервале (-∞, 0) и возрастают на интервале (0, +∞) ')
```

```python
#Задача 2
Lst = [[1, 0], [0, 0], [0, 1], [0, 2], [1, -1], [2, -1]]
count = 0
color = [0, 'r', 'orange', 'g', 'c', 'b', 'm', 'k', 'y']
fig, axes = plt.subplots(figsize = (10, 10))
plt.ylim(-20, 30) 
plt.xlim(-5, 5)
for elem in Lst:
    count += 1

    axes.plot(x1, f(x1, elem[0], elem[1]), color[count])
    axes.plot(x2, f(x2, elem[0], elem[1]), color[count], label=r'$\alpha=$'+str(elem[0])+r'$,\beta=$'+str(elem[1]))
    axes.set_xlabel('$𝑥$', fontsize = 30)
    axes.set_ylabel('$𝑓(𝑥)$', fontsize = 30)
    axes.set_title('График функции '  r'$𝑓(𝑥) =\frac{𝑥^\beta+\alpha^\beta}{𝑥^\beta}$' ' при различных ' r'$\alpha$' ' и ' r'$\beta$', fontsize = 30)
    axes.legend(fontsize = 16)
plt.savefig('2-1.svg')
plt.show()
print(u'Выводы:')
print(u'1. Графики функции при заданных значениях коэффициентов линейны')
print(u'2. При всех значениях, кроме пары α=0,β=0 графики пересекаются в точке (0,1)')
```
