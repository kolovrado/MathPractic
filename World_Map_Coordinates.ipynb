```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd

fig = plt.figure(figsize=(20, 10))
with open("world.png", 'rb') as map_file:
        map_img = plt.imread(map_file)  
axes = fig.add_axes([0.01, 0, 1, 1])
axes.imshow(map_img, zorder=0, extent=[-180, 180, -90, 90])
plt.axis('off')
df = pd.read_csv('cities.csv', encoding='utf-8')

count = 0
colors = dict(zip(set(df['Country']), ['r', 'b', 'k', 'y', 'c', 'm', 'orange', 'g'])) 

for i in range(13):
    x = df['Latitude deg'][count] + (df['Latitude min'][count] + df['Latitude sec'][count]/60)/60
    y = df['Longitude deg'][count] + (df['Longitude min'][count] + df['Longitude sec'][count]/60)/60
    col = colors[df['Country'][count]]
    print(df['Name'][count], df['Population'][count])
    if df['Population'][count] >= 1000000:
        size = df['Population'][count]/270000
        if df['Capital'][count] == 1:    
            plt.plot(x*np.cos(y/180), y, color='black', markerfacecolor=col, marker='*', markersize = size)
        else:
            plt.plot(x*np.cos(y/180), y, color = 'black', markerfacecolor=col, marker='o', markersize = size, )
        
    count += 1  
    
plt.show()
```
